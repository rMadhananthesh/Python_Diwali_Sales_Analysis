```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mp
import seaborn as sb
```

```python
df=pd.read_csv("Diwali Sales Data.csv",encoding='unicode_escape')
```

```python
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df["Amount"]=df["Amount"].astype('int')
```

```python
df['Amount'].dtype
```

```python
df.columns
```

```python
df.rename(columns={'Ages':'Age'},inplace=True)
```

```python
df.columns
```

```python
df[['User_ID', 'Age']].describe()
```

```python
a=sb.countplot(x='Gender',data=df,hue='Gender')
for x in a.containers:
    a.bar_label(x)
```

```python
sal=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sal
```

```python
v=sb.barplot(x='Gender',y='Amount',data=sal,hue='Gender')
```

```python
b=sb.countplot(x='Age Group',data=df,hue='Gender')
for y in b.containers:
    b.bar_label(y)
```

```python
t=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(['Age Group'],ascending=True)
```

```python
t
```

```python
sb.barplot(x='Age Group',y='Amount',data=t,hue='Age Group')
```

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(['Orders'],ascending=False).head(10)
```

```python
sales_state
```

```python
sb.set(rc={'figure.figsize':(16,5)})
sb.barplot(x='State',y='Orders',data=sales_state,hue='Orders')
```

*From above graphs we can see that most of the orders are from
Uttar Pradesh,Maharastra,Karnataka respectively*

Marital_Status

```python
ax=sb.countplot(x='Marital_Status',data=df,hue='Marital_Status')
sb.set(rc={'figure.figsize':(7,5)})
for x in ax.containers:
    ax.bar_label(x)
```

```python
M_a=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(['Amount'],ascending=False)
```

```python
sb.barplot(x='Marital_Status',data=M_a,hue='Gender',y='Amount')
sb.set(rc={'figure.figsize':(8,6)})
```

From the above graphs we can see that most of the buyers are married women and they have high purchasing rate

*Occupation*

```python
ax=sb.countplot(data=df,x="Occupation",hue='Occupation')
sb.set(rc={'figure.figsize':(20,5)})
for x in ax.containers:
    ax.bar_label(x)
```

```python
Occ_Am=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(['Amount'],ascending=False)
```

```python
sb.barplot(x='Occupation',y='Amount',data=Occ_Am,hue='Occupation')
```

From the above the graphs we can see that most of the buyers are working at IT Sector, HealthCare and Aviation

*Product Category*

```python
ax=sb.countplot(data=df,x="Product_Category",hue='Product_Category')
sb.set(rc={'figure.figsize':(20,5)})
for x in ax.containers:
    ax.bar_label(x)
```

```python
Prod_Am=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(['Amount'],ascending=False).head(10)
```

```python
sb.barplot(x='Product_Category',y='Amount',data=Prod_Am,hue='Product_Category')
sb.set(rc={'figure.figsize':(20,5)})
```

From the above graphs we can see that most of the sold products are from Food,Footwear and Electronics category

```python
Sale_state=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(['Orders'],ascending=False).head(10)
```

```python
sb.barplot(x='Product_ID',y='Orders',data=Sale_state,hue='Orders')
sb.set(rc={'figure.figsize':(20,5)})
```

*Conclusion*

Married Women age group 26-35 years from UP,Maharastra and Karnataka working in IT,HealthCare and Aviation are more likely to buy products from Food, Clothing and Electronics category
